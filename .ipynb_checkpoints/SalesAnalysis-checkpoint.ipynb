#### Import Necessary Python Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

### Loading And Overview Data

###### Merge data from each month into one CSV

```python
path = "sales_data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path + "/" + file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data_copy.csv", index=False)
    
```

#### Reading Merged dataframe

```python
df = pd.read_csv("all_data_copy.csv")
df.head()
```

```python
#  Shape of the Dataset
df.shape
```

```python
# DataFrame Info
df.info()
```

```python
# Statistical Summary of Numeric Columns
df.describe()
```

### Cleaning Dataset

```python
# checking missing value
missing_value = pd.DataFrame({
    'total_null_value':df.isnull().sum(), 
    'missing_value_%':df.isnull().mean()*100
})
missing_value
```

```python
# droping missing value
df = df.dropna()
df.isnull().sum().sum()
```

```python
# checking duplicated value
df.duplicated().sum()
```

```python
# droping duplicated value
df = df.drop_duplicates()
```

```python
# getting data types of columns
df.dtypes
```

```python
# conversion of data types

df = df.drop(df[df['Order Date'] == 'Order Date'].index, axis=0)

df['Order Date'] = pd.to_datetime(df['Order Date'], format="%m-%d-%y")
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
df.dtypes
```

### Featue Engineering

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df['City'] = df['Purchase Address'].apply(lambda x: f'{x.split(",")[1].strip()} ({x.split(",")[2][0:3].strip()})')
df['State'] = df['Purchase Address'].apply(lambda x: x.split(",")[2][0:3].strip())

# Reorder columns: keep first 4 columns, then add 'Sales', and then the rest excluding any duplicates
columns = list(df.columns[:4]) + ['Sales'] + [col for col in df.columns if col not in df.columns[:4] and col != 'Sales']
df = df[columns]
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day Name'] = df['Order Date'].dt.day_name()
df['Hour'] = df['Order Date'].dt.hour
df.head()
```

### Sales Trends Over Time

```python
df1 = df[['Year','Month','Sales']].copy()

# Combine Year and Month into a datetime column
df1['Date'] = pd.to_datetime(df1[['Year', 'Month']].assign(DAY=1))

#figure size
plt.figure(figsize=(17,5))

#year
plt.subplot(1, 3, 1)
sns.lineplot(data=df1, x='Date', y='Sales', estimator='sum')
plt.ticklabel_format(style='plain', axis='y')
# Format y-axis to show in millions
formatter = FuncFormatter(lambda x, _: f'{x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)
plt.xticks(rotation=45)
plt.ylabel('($) Total Sales')
plt.title('Yearly Sales Performance', fontsize=14, fontweight='bold')

#month
plt.subplot(1, 3, 2)
sns.lineplot(data=df, x = 'Month', y = 'Sales', estimator='sum')
#plt.ticklabel_format(style='plain', axis='y') for plain text
# Format y-axis to show in millions
formatter = FuncFormatter(lambda x, _: f'{x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)
plt.xticks(rotation=45)
plt.ylabel('($) Total Sales')
plt.title('Monthly Sales Seasonality', fontsize=14, fontweight='bold')


#wekday
plt.subplot(1, 3, 3)
sns.lineplot(data=df, x = 'Day Name', y = 'Sales', estimator='sum')
#plt.ticklabel_format(style='plain', axis='y') for plain text
# Format y-axis to show in millions
formatter = FuncFormatter(lambda x, _: f'{x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)
plt.xticks(rotation=45)
plt.ylabel('($) Total Sales')
plt.title('Sales by Day of the Week', fontsize=14, fontweight='bold')

plt.suptitle(
    "Sales Trends Over Time: Yearly, Monthly, and Weekly Insights",
    fontsize=22, fontweight='bold', color='darkgreen'
)
# plt.figtext(
#     0.5, 1.02,
#     "Analyzing patterns in total sales across years, months, and weekdays",
#     ha='center', fontsize=12, color='gray'
# ) for description

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Give room for the suptitle
plt.show()
```

#### Question 2: What products are most often sold together?

```python
from itertools import combinations
from collections import Counter


# Step 1: Drop duplicates (in case of duplicate entries)
df = df.drop_duplicates(subset=['Order ID', 'Product'])

# Step 2: Group products by each order
grouped = df.groupby('Order ID')['Product'].apply(lambda x: ', '.join(sorted(x))).reset_index()

# Step 3: Find all product pairs sold together
combo_list = []

for products in grouped['Product']:
    items = products.split(', ')
    if len(items) > 1:
        combo_list += combinations(items, 2)

# Step 4: Count frequency of each pair
pair_counts = Counter(combo_list)

# Step 5: Convert to DataFrame for plotting
top_pairs = pair_counts.most_common(10)
top_pairs_df = pd.DataFrame(top_pairs, columns=['Pair', 'Count'])
top_pairs_df['Pair Label'] = top_pairs_df['Pair'].apply(lambda x: f"{x[0]} + {x[1]}")

# Step 6: Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=top_pairs_df, x='Count', y='Pair Label', palette='viridis', hue='Pair Label')
plt.title("Top 10 Product Pairs Sold Together", fontsize=16, fontweight='bold')
plt.xlabel("Number of Orders")
plt.ylabel("Product Pair")
plt.tight_layout()
plt.show()
```
